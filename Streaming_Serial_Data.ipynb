# **Streaming Arduino Serial Data to CSV**

```python
import serial
import time
```

```python
board = serial.Serial('COM7', 115200, timeout=1)

# File is overwritten
fhand = open('electrical_data.csv', 'w')
```

```python
try:
    while True:
        line = board.readline()
        if line:
            line = line.decode()
            fhand.write(line)
            print(line)

except KeyboardInterrupt:

    # Always close the serial port; otherwise, there will be problems when opening it the next time.
    # Not closing the file handle can lead to data loss.
    board.close()
    fhand.close()
    print('Serial port and file handle closed.')
```
